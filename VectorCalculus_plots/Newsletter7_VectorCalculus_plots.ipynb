# Newsletter 7. Vector Calculus: Flux integrals and the Divergence Theorem

## Libraries and main settings

```python
#Numerical computation
import numpy as np

#graph library
import matplotlib.pyplot as plt

#3d frame 
from mpl_toolkits.mplot3d import Axes3D

#To recognize LaTeX commands
plt.rc('text', usetex=True)

#font family
plt.rc('font', family='serif')

#style sheet
plt.style.use('seaborn-paper')

#change the background colour 
c_background = '#FFFFFF'

#change the XYZ planes colour
xyz_background = (250/255, 249/255, 248/255)
```

## Plot 1. Flux through a cylinder

```python
#for non interactive plot
%matplotlib inline

# ------------------- CYLINDER ----------------------------
#Parameters
theta = np.linspace( 0, 2*np.pi, 100 )
l = np.linspace( 0, 20, 100 )

theta,l = np.meshgrid(theta,l)

#vector components of \vec{X} ellipse
X_cylinder = 15*np.cos(theta)
Y_cylinder = 15*np.sin(theta)
Z_cylinder = l

# ----------------- VECTOR FIELD --------------------------
#position of arrows in the axes
X = np.linspace( -20, 20, 8 )
Y = np.linspace( -20, 20, 8 )
Z = np.linspace( 0*1, 20, 8 )

#create meshgrid
X, Y, Z = np.meshgrid( X, Y, Z )

#value of the vector field
U = np.log(X**2 + Y**2) * X
V = np.log(X**2 + Y**2) * Y
W = Z*0

#----------------------------------- PLOTS-------------------------


#Create figure and the background color
fig = plt.figure( figsize=(10.5,6),
                 dpi = 200, #make your plot and HD-Plot                 
                )
ax = Axes3D(fig)
ax.set_facecolor( c_background )
ax.w_xaxis.set_pane_color( xyz_background )
ax.w_yaxis.set_pane_color( xyz_background )
ax.w_zaxis.set_pane_color( xyz_background )


#Surface
ax.plot_surface( X_cylinder, Y_cylinder, Z_cylinder,
                shade = False,
                color = '#D1AF94',
                alpha = 1.0,
                cstride = 1,
                rstride = 1,
               )

#create the quiver
q = ax.quiver(X, Y, Z, 
              U, V, W, 
              length= 2, 
              lw = 1.5,
              normalize = True,
              colors = '#6A5BA3',
              label = r'Campo vectorial $\vec{V}(x,y,z)$'
             )

#label of our axes
ax.set_xlabel( '$x$ \n[cm]', 
              labelpad = 20,
              size = 20
             )
ax.set_ylabel( '$y$ \n[cm]', 
              labelpad = 20,
              size = 20
             )
ax.set_zlabel( '$z$ \n[cm]', 
              labelpad = 15,
              size = 20
             )

#size of the axes
ax.tick_params( axis = 'x', labelsize = 15 )
ax.tick_params( axis = 'y', labelsize = 15 )
ax.tick_params( axis = 'z', labelsize = 15 )

ax.legend(loc=0, prop = {'size':15})
ax.grid(b = False)

plt.show()
```

## Plot 3. Small portion of the cylinder (a plane) $+$ normal unit vectors of the surface $\hat{n}$ $+$ vector field $\vec{V}(x,y,z)$

```python
#for non interactive plot
%matplotlib inline

# ------------------- CYLINDER ----------------------------
#Parameters
theta = np.linspace( (6/4)*np.pi, 2*np.pi ,30 )
l = np.linspace( 0, 5, 30 )

theta,l = np.meshgrid(theta,l)

#vector components of \vec{X} ellipse
X_cylinder = 15*np.cos(theta)
Y_cylinder = 15*np.sin(theta)
Z_cylinder = l

# ----------------- VECTOR FIELD --------------------------
#position of arrows in the cylinder
#first half of parameters
theta_V = np.linspace( (7/4)*np.pi, 2*np.pi, 6 )
l_V = np.linspace( 0, 5, 6 )

#create meshgrid
theta_V, l_V = np.meshgrid( theta_V, l_V )

X = 15*np.cos(theta_V)
Y = 15*np.sin(theta_V)
Z = l_V


#value of the vector field
U = np.log(X**2 + Y**2) * X
V = np.log(X**2 + Y**2) * Y
W = Z*0

#------------------------ NORMAL VECTORS OF THE CYLINDER
#position of the vectors
#first half of parameters
theta_n = np.linspace( (6/4)*np.pi, (7/4)*np.pi, 6 )
l_n = np.linspace( 0, 5, 6 )

#create meshgrid
theta_n, l_n = np.meshgrid( theta_n, l_n )

X_n = 15*np.cos(theta_n)
Y_n = 15*np.sin(theta_n)
Z_n = l_n

#value of the vectors
U_n = X_n/15
V_n = Y_n/15
W_n = Z_n*0

#----------------------------------- PLOTS-------------------------
#Create figure and the background color
fig = plt.figure( figsize=(10.5,6),
                 dpi = 200, #make your plot and HD-Plot                 
                )
ax = Axes3D(fig)
ax.set_facecolor( c_background )
ax.w_xaxis.set_pane_color( xyz_background )
ax.w_yaxis.set_pane_color( xyz_background )
ax.w_zaxis.set_pane_color( xyz_background )


#Surface
ax.plot_surface( X_cylinder, Y_cylinder, Z_cylinder,
                shade = True,
                color = '#B39887',
                alpha = 1.0,
                cstride = 7,
                rstride = 35,
               )

#normal vectors
q_n = ax.quiver( X_n + 0.1 , Y_n+0.1, Z_n+0.1,
                U_n, V_n, W_n,
                colors = '#64A0B9',
                length = 1.5,
                lw = 1,
                label = r'Vector normal $\hat{n}$'
         )

#create the vector field
q = ax.quiver(X, Y, Z, 
              U, V, W, 
              length= 2 , 
              lw = 1,
              normalize = True,
              colors = '#C0B3EE',
              label = r'Campo vectorial $\vec{V}(x,y,z)$'
             )

#------------------other settings

#label of our axes
ax.set_xlabel( '$x$ \n[cm]', 
              labelpad = 20,
              size = 20
             )
ax.set_ylabel( '$y$ \n[cm]', 
              labelpad = 20,
              size = 20
             )
ax.set_zlabel( '$z$ \n[cm]', 
              labelpad = 15,
              size = 20
             )

#size of the axes
ax.tick_params( axis = 'x', labelsize = 15 )
ax.tick_params( axis = 'y', labelsize = 15 )
ax.tick_params( axis = 'z', labelsize = 15 )


ax.legend(loc=0, prop = {'size':15})
ax.grid(b = False)

plt.show()
```

## Flux Integral

\begin{equation*}
\iint_{S} \vec{F}\left( \vec{X}(u,v) \right)\cdot \left( \vec{X}_{u}\times\vec{X}_{v} \right) \text{ d}u\text{d}v
\end{equation*}

## Plot 4. Flux through a hemisphere

```python
#for non interactive plot
%matplotlib inline

# ------------------- SPHERE ----------------------------
#Parameters
theta = np.linspace( 0, 2*np.pi, 100 )
phi = np.linspace( 0, 0.5*np.pi , 100 )

theta,phi = np.meshgrid(theta, phi)

#vector components of \vec{X} sphere
X_sphere = 3*np.cos(theta)*np.sin(phi)
Y_sphere = 3*np.sin(theta)*np.sin(phi)
Z_sphere = 3*np.cos(phi)

# ----------------- VECTOR FIELD --------------------------
#position of arrows in the axes
X = np.linspace( -3.5, 3.5, 8 )
Y = np.linspace( -3.5, 3.5, 8 )
Z = np.linspace( -0.0, 3.5, 8 )

#create meshgrid
X, Y, Z = np.meshgrid( X, Y, Z )

#value of the vector field
U = X
V = Y
W = Z**4

#----------------------------------- PLOTS-------------------------

#Create figure and the background color
fig = plt.figure( figsize=(10.5,6),
                 dpi = 200, #make your plot and HD-Plot                 
                )
ax = Axes3D(fig)
ax.set_facecolor( c_background )
ax.w_xaxis.set_pane_color( xyz_background )
ax.w_yaxis.set_pane_color( xyz_background )
ax.w_zaxis.set_pane_color( xyz_background )



#Surface
ax.plot_surface( X_sphere, Y_sphere, Z_sphere,
                color = '#C6B1AB',
                alpha = 1.0,
                cstride = 1,
                rstride = 1,
                shade = False,
               )

#create the quiver
q = ax.quiver(X, Y, Z, 
              U, V, W, 
              length= 0.5 , 
              lw = 1,
              normalize = True,
              colors = '#6A5BA3',
              label = r'Campo vectorial $\vec{V}(x,y,z)$',
             )



#label of our axes
ax.set_xlabel( '$x$ \n[m]', 
              labelpad = 20,
              size = 20
             )
ax.set_ylabel( '$y$ \n[m]', 
              labelpad = 20,
              size = 20
             )
ax.set_zlabel( '$z$ \n[m]', 
              labelpad = 15,
              size = 20
             )

#size of the axes
ax.tick_params( axis = 'x', labelsize = 15 )
ax.tick_params( axis = 'y', labelsize = 15 )
ax.tick_params( axis = 'z', labelsize = 15 )


ax.legend(loc=0, prop = {'size':15})
ax.grid(b = False)


plt.show()
```

## The Divergence Theorem

\begin{equation*}
\iint \vec{F}\cdot\hat{n}\text{ d}S = \iiint\limits_{R} \left( \vec{\nabla}\cdot\vec{F} \right)\text{d}V
\end{equation*}
