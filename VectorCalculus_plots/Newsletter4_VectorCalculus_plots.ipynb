## Divergence

### Plot 1. Vector field $\vec{F}(x,y) = \left( x^{2} - y^{2} \right)\hat{i} + \left( 2xy \right)\hat{j}$

```python
#non interactive plots
%matplotlib inline

#Domain of the vector field
X = np.linspace( -10, 10, 26 )
Y = np.linspace( -10, 10, 26 )

#Points in the XY plane
X,Y = np.meshgrid(X,Y)

#Value each component of the vector field => F(x,y) = u*i + v*j
U = X**2 - Y**2
V = 2*X*Y

#magnitude of all the vectors
Magnitude = np.sqrt(U**2 + V**2)

#Normalize  so all of them have magnitude of one
U = U/Magnitude
V = V/Magnitude

#----------------------------------------- PLOT TIME ---------------------------------------------

#Create figure and dimensions
plt.figure(figsize=(10.6,6), 
           dpi = 200,
          )

#Create the vector plot
plt.quiver( X, Y, #Position of each vector
           U, V, #value of each vector
           Magnitude, #magnitude of each vector, this will help the cmap
           width = 0.002,
           headwidth = 4, 
           headlength = 6,
           cmap = plt.cm.PuBuGn_r #assigns a color to each vector depending its magnitude
          )

#Add the title and the labels of the axes
plt.title(r'$\vec{F}(x,y) = \left( x^{2} - y^{2} \right) \ \hat{i} + \left( 2xy \right) \ \hat{j}$', 
          size = 25, 
          pad = 25)
plt.xlabel( r'$x$', size = 20 )
plt.ylabel( r'$y$', size = 20 )

#change size of the tick params
plt.tick_params( labelsize = 15 )

#create a color bar for the vector field.
#the color bar shows us the color assigned to the magnitude of the vector 
color_bar = plt.colorbar( orientation = 'vertical', pad = 0.1 )

#Set a title to the colorbar
color_bar.set_label( label = 'Magnitud del vector', size = 20 )

#change the size of the numbers in the colorbar
color_bar.ax.tick_params( labelsize = 15 )


plt.show()
```

### Plot 4. Rotational value of each point in the vector field

**Vector field**
$$\vec{F}(x,y) = \cos( x+y ) \ \hat{i} + \sin(xy) \ \hat{j}$$

**Rotational of the vector field**
$$\vec{\nabla}\times\vec{F} = \frac{\partial F_{y}}{\partial x} - \frac{\partial F_{x}}{\partial y} =y \cos{\left(x y \right)} + \sin{\left(x + y \right)} = g(x,y)$$

```python
#non interactive plots
%matplotlib inline

#-------------------------- SCALAR FIELD FOR THE ROTATIONAL -----------------------------
#domain of the rotational
X_rot = np.linspace( -5, 5, 100 )
Y_rot = np.linspace( -5, 5, 100 )

#Points in the XY plane
X_rot, Y_rot = np.meshgrid( X_rot, Y_rot )

#scalar function for the values of the vector field
g_rot = Y_rot*np.cos( X_rot*Y_rot ) + np.sin( X_rot + Y_rot )


#---------------------------- VECTOR FIELD -----------------------------------------------
#Domain of the vector field
X = np.linspace( -5, 5, 36 )
Y = np.linspace( -5, 5, 36 )

#Points in the XY plane
X,Y = np.meshgrid(X,Y)

#Value each component of the vector field => F(x,y) = u*i + v*j
U = np.cos(X + Y)
V = np.sin(X*Y)

#magnitude of all the vectors
Magnitude = np.sqrt(U**2 + V**2)

#Normalize  so all of them have magnitude of one
U = U/Magnitude
V = V/Magnitude

#----------------------------------------- PLOT TIME ---------------------------------------------

#Create figure and dimensions
plt.figure(figsize=(10.6,6), 
           dpi = 200,
          )

#create contourf
plt.contourf(X_rot, Y_rot, g_rot, 
             cmap=plt.cm.RdYlBu_r,
             levels = 30,
            )

#create a color bar for the vector field.
#the color bar shows us the color assigned to the magnitude of the vector 
color_bar = plt.colorbar( orientation = 'vertical', pad = 0.1 )

#Set a title to the colorbar
color_bar.set_label( label = 'Valor del rotacional', size = 20 )

#change the size of the numbers in the colorbar
color_bar.ax.tick_params( labelsize = 15 )

#Create the vector plot
plt.quiver( X, Y, #Position of each vector
           U, V, #value of each vector
           #Magnitude, #magnitude of each vector, this will help the cmap
           color = 'black',
           width = 0.002,
           headwidth = 3, 
           headlength = 5,
           #cmap = plt.cm.PuBu_r #assigns a color to each vector depending its magnitude
          )

#Add the title and the labels of the axes
#plt.title(r'$\vec{F}(x,y) = \cos( x+y ) \ \hat{i} + \sin(xy) \ \hat{j}$', 
#          size = 25, 
#          pad = 25)
plt.xlabel( r'$x$', size = 20 )
plt.ylabel( r'$y$', size = 20 )

#change size of the tick params
plt.tick_params( labelsize = 15 )


plt.show()
```

## Applications: Differencial Equations 

### Plot 5. Phase space of $\ddot{\theta}(t) = - \mu\dot{\theta}(t) - \frac{g}{L}\sin\left( \theta (t)  \right) $

```python
#non interactive plots
%matplotlib inline

#damping variable
mu = 0.2

#gravity
g = 9.8

#length
L = 10

#---------------------------- VECTOR FIELD -----------------------------------------------
#Domain of the vector field
X = np.linspace( -2*np.pi, 2*np.pi, 30 )
Y = np.linspace( -2*np.pi, 2*np.pi, 30 )

#Points in the XY plane
X,Y = np.meshgrid(X,Y)

#Value each component of the vector field => F(x,y) = u*i + v*j
U = Y
V = -mu*Y - (g/L)*np.sin(X)

#magnitude of all the vectors
Magnitude = np.sqrt(U**2 + V**2)

#Normalize  so all of them have magnitude of one
U = U/Magnitude
V = V/Magnitude

#----------------------------------------- PLOT TIME ---------------------------------------------

#Create figure and dimensions
plt.figure(figsize=(10.6,6), 
           dpi = 200,
          )


#Create the vector plot
plt.quiver( X, Y, #Position of each vector
           U, V, #value of each vector
           Magnitude, #magnitude of each vector, this will help the cmap
           width = 0.002,
           headwidth = 3, 
           headlength = 5,
           cmap = plt.cm.PuBuGn_r #assigns a color to each vector depending its magnitude
          )

#create a color bar for the vector field.
#the color bar shows us the color assigned to the magnitude of the vector 
color_bar = plt.colorbar( orientation = 'vertical', pad = 0.1 )

#Set a title to the colorbar
color_bar.set_label( label = 'Magnitud del vector', size = 20 )

#change the size of the numbers in the colorbar
color_bar.ax.tick_params( labelsize = 15 )


#Add the title and the labels of the axes
plt.title(r'Espacio fase de $\ddot{\theta}(t) = - \mu\dot{\theta}(t) - \frac{g}{L}\sin\left( \theta (t)  \right)$', 
          size = 25, 
          pad = 10)
plt.xlabel( r'$x$', size = 20 )
plt.ylabel( r'$y$', size = 20 )

#change size of the tick params
plt.tick_params( labelsize = 15 )


plt.show()
```
