## Libraries and main settings

```python
#Numerical computation
import numpy as np

#graph library
import matplotlib.pyplot as plt

#3d frame 
from mpl_toolkits.mplot3d import Axes3D 

#To recognize LaTeX commands
plt.rc('text', usetex=True)

#font family
plt.rc('font', family='serif')

#style sheet
plt.style.use('seaborn-poster')
```

## Plot 1. 3D Parametric curve

$$\vec{x}(t) = 4\sin(2t) \  \hat{i} + 4\cos(2t) \  \hat{j} + \frac{t}{8} \  \hat{k} \\ 0\leq t \leq 2\pi$$

```python
#Interactive 3D plot
%matplotlib notebook

#Create the parameter's domain
t_line = np.linspace(-1, 2 * np.pi, 200)

#Define the function of every component
x = 4*np.sin(2*t_line)
y = 4*np.cos(2*t_line)
z = t_line/8

#Create a figure
fig = plt.figure()

#Create 3D axes
ax = fig.add_subplot(111, projection='3d')

#Plot the 3d parametric curve
ax.plot(x, y, z, 
        linewidth = 1.5,
        color = '#F53A86')

#Add title
ax.set_title( r'$\vec{x}(t) = 4\sin(2t) \  \hat{i} + 4\cos(2t) \  \hat{j} + \frac{t}{8} \  \hat{k}$',
             size = 25,
            pad = 10 )

#Add label's name and the pad
ax.set_xlabel( r'$x$', size = 20, labelpad = 15 )
ax.set_ylabel( r'$y$', size = 20, labelpad = 15 )
ax.set_zlabel( r'$z$', size = 20, labelpad = 15 )

#remove the grid so it looks better (if you want)
ax.grid( b = False )

plt.tight_layout()
plt.show()
```

## Plot 2. Parametric curve: XY-plane view

```python
#In line plots - non interactive plots
%matplotlib inline

#Create a figure
plt.figure( figsize=(8,8) )

#Plot the the x and y components of the parametric curve
plt.plot(x, y, 
         color = '#F53A86', 
         linewidth = 1.5)

#Add title
plt.title(r'Vista en el plano $XY$ de $\vec{x}(t)$', 
          pad = 15 )
#Add label's name
plt.xlabel(r'$\hat{i}$')
plt.ylabel(r'$\hat{j}$')

plt.tight_layout()
plt.show()
```

## Plot 3. Parametric curve: XZ-plane view

```python
#In line plots - non interactive plots
%matplotlib inline

#Create a figure
plt.figure( figsize=(8,8) )

#Plot the the x and y components of the parametric curve
plt.plot(x, z, 
         color = '#F53A86', 
         linewidth = 1.5)

#Add title
plt.title(r'Vista en el plano $XY$ de $\vec{x}(t)$', 
          pad = 15 )
#Add label's name
plt.xlabel(r'$\hat{i}$')
plt.ylabel(r'$\hat{k}$')

plt.tight_layout()
plt.show()
```
