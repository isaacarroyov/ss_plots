## Libraries and main settings

```python
#Numerical computation
import numpy as np

#graph library
import matplotlib.pyplot as plt

#3d frame 
from mpl_toolkits.mplot3d import Axes3D

#To recognize LaTeX commands
plt.rc('text', usetex=True)

#font family
plt.rc('font', family='serif')

#style sheet
plt.style.use('seaborn-poster')
```

## Vector Fields

### Plot 1. StreamPlot

$$\vec{F}(x,y) = \cos\left( \sin(y) \right) \ \hat{i} + \cos(x) \ \hat{j}$$

```python
#in line plots, non interactive plots
%matplotlib inline

#Domain of the vector field
X = np.linspace(-10,10,35)
Y = np.linspace(-10,10,35)

#Points in the XY plane
X,Y = np.meshgrid( X, Y )

#vector associated with each point
U = np.cos( np.sin(Y) )
V = np.cos( X )

#Create figure and dimensions
plt.figure( figsize = (4,4),
           dpi = 200 #make your plot an HDPlot, remove it if you're not in a jupyter notebook
          ) 

#make the streamplot
plt.streamplot( X, Y, #position of the vector
               U, V, #value of the vector
               color = '#15A7DF', #General color of the streamplot
               arrowsize = 1, 
               arrowstyle='fancy',
               linewidth = 0.75
              )

#add text to the labels
plt.xlabel( '$x$', size = 15 )
plt.ylabel( '$y$', size = 15 )

#change the size of the tick params
plt.tick_params( labelsize = 10 )

#set the plot's "region"
plt.xlim( -10,10 )
plt.ylim( -10,10 )

plt.tight_layout()
plt.show()
```

### Plot 2. Diverging vector field with a colorbar

$$\vec{F}(x,y) = x \ \hat{i} + y \ \hat{j}$$

[Colormaps](https://matplotlib.org/tutorials/colors/colormaps.html)

```python
#in line plots, non interactive plots
%matplotlib inline

#Domain of the vector field
X = np.linspace(-10,10,24)
Y = np.linspace(-10,10,24)

#Points in the XY plane
X,Y = np.meshgrid( X, Y )

#vector associated with each point
U = X
V = Y

#Magnitude of the vectors
Magnitude = np.sqrt( U**2 + V**2 )

#Normalize the vectors to make every vector the same lenght 
U = U/Magnitude
V = V/Magnitude

#Create figure and dimensions
plt.figure( figsize = (10,8),
           dpi = 200 #make your plot an HDPlot, remove it if you're not in a jupyter notebook
          )
#add the vector plot (quiver)
plt.quiver( X, Y, #position of each vector
           U, V, #direction of each vector
           Magnitude, #This will help the cmap 
           cmap = plt.cm.summer , #assigns a color to each vector depending its magnitude 
           width = 0.002, #size of the arrows 
           headwidth = 3,
           headlength = 5,
          )

plt.xlabel( '$x$', size = 25 )
plt.ylabel( '$y$', size = 25 )
plt.tick_params( labelsize = 15 )
plt.xlim( -10,10 )
plt.ylim( -10,10 )

#create a color bar for the vector field.
#the color bar shows us the color assigned to the magnitude of the vector 
colorbar =  plt.colorbar()

#change the size of the numbers in the colorbar
colorbar.ax.tick_params( labelsize = 15 )

#Set a title to the colorbar
colorbar.set_label( label = 'Magnitude of the vector', size = 25 )

plt.tight_layout()
plt.show()
```

### Plot 3. Vector field with horizontal colorbar

$$\vec{F}(x,y) = \left( y \right) \ \hat{i} + \left( x^{2}\right) \ \hat{j} $$

```python
#in line plots, non interactive plots
%matplotlib inline

#Domain of the vector field
X = np.linspace(-5, 5, 24 )
Y = np.linspace(-5, 5, 24 )

#Points in the XY plane
X,Y = np.meshgrid( X, Y )

#vector associated with each point
U = Y
V = X**2

#Magnitude of the vectors
Magnitude = np.sqrt( U**2 + V**2 )

#Normalize the vectors to make every vector the same lenght 
U = U/Magnitude
V = V/Magnitude

#Create figure and dimensions
plt.figure( figsize = (8,10),
           dpi = 200 #make your plot an HDPlot, remove it if you're not in a jupyter notebook
          )

#add the vector plot (quiver)
plt.quiver( X, Y, #position of each vector
           U, V, #direction of each vector
           Magnitude, #This will help the cmap 
           cmap = plt.cm.summer , #assigns a color to each vector depending its magnitude 
           width = 0.002, #size of the arrows 
           headwidth = 3,
           headlength = 5,
          )

plt.xlabel( '$x$', size = 25 )
plt.ylabel( '$y$', size = 25 )
plt.tick_params( labelsize = 15 )


#to position the color bar horizontally just add orientation = 'horizontal', the default value is 'vertical'
colorbar = plt.colorbar( orientation = 'horizontal', pad = 0.15 )
colorbar.ax.tick_params( labelsize = 15 ) 
colorbar.set_label( label = 'Magnitude of the vector', size = 25 )

plt.tight_layout()
plt.show()
```

## Directional Derivative

### Plot 4. Scalar function

$$f(x,y) = z = -2x^{2} - 3y^{2} + 5000$$

```python
#interactive plots
%matplotlib notebook

#Domain of the function
X = np.linspace(-5, 5, 500 )
Y = np.linspace(-5, 5, 500 )

#Points in the XY plane
X,Y = np.meshgrid( X, Y )

# f(x,y) = z
Z = -2*(X**2) - 3*(Y**2) + 5000

#Create figure and dimensions
fig = plt.figure( 
                 #figsize=(9,5), if you want an inline plot, if not, comment on it
                 #dpi = 180 #make your plot an HDPlot, remove it if you're not in a jupyter notebook
                )

#Inside the figure, make axes in 3D
ax = Axes3D(fig)

#Create the "mountain" plot
ax.plot_surface(X, Y, Z, cmap = plt.cm.YlOrBr_r)

#modify the label of the axes and their label pad
ax.set_xlabel('$x$', size = 25, labelpad = 15) #labelpad => separates the label from the axis
ax.set_ylabel('$y$', size = 25, labelpad = 15)
ax.set_zlabel('Altura', size = 25, labelpad = 20)

#modify the tick params of each axis
ax.tick_params( axis = 'x', labelsize = 15)
ax.tick_params( axis = 'y', labelsize = 15)
ax.tick_params( axis = 'z', labelsize = 15)

#remove the grid if you want (b=False), if not => b = True
ax.grid(b=False)

plt.show()
```

### Plot 5. Contour lines of the function

$$f(x,y) = z = -2x^{2} - 3y^{2} + 5000$$

```python
#in line plots => non interactive plots
%matplotlib inline

#Domain of the function
X = np.linspace(-5, 5, 500 )
Y = np.linspace(-5, 5, 500 )

#Points in the XY plane
X,Y = np.meshgrid( X, Y )

# f(x,y) = z
Z = -2*(X**2) - 3*(Y**2) + 5000

#Create figure and dimensions
plt.figure( figsize=(10.5,6),
           #dpi = 180, #make your plot an HDPlot, remove it if you're not in a jupyter notebook
          )

#add the contour lines with contour
plt.contour( X, Y, Z,
            levels = 20, #how many lines do you want
            cmap = plt.cm.YlOrBr_r,
            linewidths = 1.5
           )

#add a title and labels to the axes
plt.title( r'Curvas de nivel de $f(x,y) = z = -2x^{2} - 3y^{2} + 5000 $', 
          size = 25,
          pad = 15,
         )
plt.xlabel( '$x$', size = 20 )
plt.ylabel( '$y$', size = 20 )

#modify the tick_params
plt.tick_params( labelsize = 15 )

#Adding a color bar to the figure
colorbar_mountain = plt.colorbar( orientation = 'vertical', pad = 0.08 )
colorbar_mountain.ax.tick_params( labelsize = 15 )
colorbar_mountain.set_label( label = '$f(x,y) =$ Altura', size = 20 )

plt.tight_layout()
plt.show()
```

### Plot 6. Contour lines + point $(x_0, y_0)$ + vector $\vec{u}$ + gradient vector $\vec{\nabla}f_{(x_0, y_0)}$ 

$$(x_0. y_0) = (2,4) \\ \vec{\nabla}f(x,y) = -4x \ \hat{i} - 6y \ \hat{j} \\ \vec{u} = -10 \ \hat{i} - 15 \ \hat{j} $$

```python
#in line plots => non interactive plots
%matplotlib inline

#----------------------- CONTOUR LINES -----------------------------------
#Domain of the function
X = np.linspace( 0, 5, 500 )
Y = np.linspace( 0, 5, 500 )

#Points in the XY plane
X,Y = np.meshgrid( X, Y )

# f(x,y) = z
Z = -2*(X**2) - 3*(Y**2) + 5000

#----------------------- Point x0,y0 -----------------------------------
x_0 = 2
y_0 = 4

#----------------------- u vector -----------------------------------
U = -10
V = -15

#----------------------- gradient vector -----------------------------------
gradx = -4*x_0
grady = -6*y_0

#----------------------- Plot -----------------------------------
#Create figure and dimensions
plt.figure( figsize=(10.5,6),
           #dpi = 180, #make your plot an HDPlot, remove it if you're not in a jupyter notebook
          )
#point x_0, y_0
plt.scatter( x_0, y_0,
            s = 50,
            color = '#B855EC',
            label = r'$P = (x_0,y_0) = (2,4)$'
           )

# u vector
plt.quiver(x_0, y_0,
           U, V,
           width = 0.005,  
           headwidth = 2, 
           headlength = 5, 
           label = r'$\vec{u} = -10 \ \hat{i} - 15 \ \hat{j}$'
          )
# grad vector
plt.quiver(x_0, y_0,
           gradx, grady,
           width = 0.005, 
           headwidth = 2,
           headlength = 5, 
           label = r'$\vec{\nabla}f_{(x_0,y_0)} = -4x_0 \ \hat{i} - 6y_0 \ \hat{j}$',
           color = '#5589EC'
          )
#Add the contourplot 
plt.contour( X, Y, Z,
            levels = 20, #how many lines do you want
            cmap = plt.cm.YlOrBr_r,
            linewidths = 1.5
           )

#set x and y labels
plt.xlabel( '$x$', size = 20 )
plt.ylabel( '$y$', size = 20 )

#modify the tick_params
plt.tick_params( labelsize = 15 )

#ad a legend
plt.legend( loc = 0, prop = {'size':15} ).get_frame().set_facecolor('#FFFFFF')

#Adding a color bar to the figure
colorbar_mountain = plt.colorbar( orientation = 'vertical', pad = 0.08 )
colorbar_mountain.ax.tick_params( labelsize = 15 )
colorbar_mountain.set_label( label = '$f(x,y) =$ Altura', size = 20 )

plt.tight_layout()
plt.show()
```

### Plot 7. Gradient field of the scalar field

```python
#in line plots => non interactive plots
%matplotlib inline

#----------------------- CONTOUR LINES -----------------------------------
#Domain of the function
X = np.linspace( -5, 5, 30 )
Y = np.linspace( -5, 5, 30 )

#Points in the XY plane
X,Y = np.meshgrid( X, Y )

# f(x,y) = z
Z = -2*(X**2) - 3*(Y**2) + 5000

#----------------- GRADIENT FIELD -------------------------------------
gradx = -4*X
grady = -6*Y

Magnitude = np.sqrt( gradx**2 + grady**2 )

gradx = gradx/Magnitude
grady = grady/Magnitude

#-------------- CONTOUR LINES ---------------------------------------
#Create figure and dimensions
fig = plt.figure( 
                 figsize=(10.5,6),
                 dpi = 180, #make your plot an HDPlot, remove it if you're not in a jupyter notebook
                )
#gradient field
plt.quiver(X, Y,
           gradx, grady,
           width = 0.002, 
           headwidth = 2,
           headlength = 5, 
           label = r'$\vec{\nabla}f_{(x_0,y_0)} = -4x_0 \ \hat{i} - 6y_0 \ \hat{j}$',
           color = 'black'
          )

#Add the contourplot 
plt.contour( X, Y, Z,
            levels = 15, #how many lines do you want
            cmap = plt.cm.YlOrBr_r,
            linewidths = 1.5
           )

plt.title(r'Curvas de nivel y gradiente $\left( \vec{\nabla} \right)f$', 
         pad = 20, size = 20)
#set x and y labels
plt.xlabel( '$x$', size = 20 )
plt.ylabel( '$y$', size = 20 )

#modify the tick_params
plt.tick_params( labelsize = 15 )

#ad a legend
#plt.legend( loc = 0, prop = {'size':15} ).get_frame().set_facecolor('#FFFFFF')

#Adding a color bar to the figure
colorbar_mountain = plt.colorbar( orientation = 'vertical', pad = 0.08 )
colorbar_mountain.ax.tick_params( labelsize = 15 )
colorbar_mountain.set_label( label = '$f(x,y) =$ Altura', size = 20 )

plt.tight_layout()
plt.show()
```
