# Newsletter 8. Vector Calculus: Stokes' Theorem

$$\oint\limits_{C}\vec{F}\cdot\text{d}\vec{r} = \iint\limits_{S}\left( \vec{\nabla}\times\vec{F} \right) \cdot \hat{n} \text{ d}S $$

## Libraries and main settings

```python
#Numerical computation
import numpy as np

#graph library
import matplotlib.pyplot as plt

#3d frame 
from mpl_toolkits.mplot3d import Axes3D

#To recognize LaTeX commands
plt.rc('text', usetex=True)

#font family
plt.rc('font', family='serif')

#style sheet
plt.style.use('dark_background')

#change the background colour 
hex_background = '#594F4F'

#change the XYZ planes colour
rgba_planes = (56/255, 49/255, 49/255)
```

[Style Sheet - Matplotlib](https://matplotlib.org/3.1.0/gallery/style_sheets/style_sheets_reference.html)

## Plot 1. Vector plot 1 $+$ Curve $C$

```python
#quiver plot
X = np.linspace(-5,5,20)
Y = np.linspace(-5,5,20)

#meshgrid 
X,Y = np.meshgrid(X,Y)

#vector values
U = 0
V = 1

Mag = np.sqrt(U**2 + V**2)

U = U/Mag
V = V/Mag

#Create figure and the background color
plt.figure( figsize=(10,10),
           dpi = 200,#make your plot an HDPlot, remove it if you're not in a jupyter notebook
           facecolor=hex_background,
                )

ax = plt.axes()
ax.set_facecolor(hex_background)


plt.quiver( X, Y,
           U, V,
           color = '#9DE0AD',
           width = 0.0035, #grosor de flechas
           headwidth = 4, #grosor de la cabeza
           headlength = 6, #finura/largo de la cabeza
           pivot = 'mid',
           scale = 30,
           label = 'Vector Field'
          )
plt.vlines( 4, -4, 4, color = '#FFFFFF', linewidth = 2.5, label = 'Curve $C(t)$' )
plt.vlines( -4, -4, 4, color = '#FFFFFF', linewidth = 2.5 )
plt.hlines( 4, -4, 4, color = '#FFFFFF', linewidth = 2.5 )
plt.hlines( -4, -4, 4, color = '#FFFFFF', linewidth = 2 )


#axes label
plt.xlabel(r'$x$', size = 20)
plt.ylabel(r'$y$', size = 20)

#size of tick params
plt.tick_params( labelsize = 15 )

#grid and legend
#plt.grid(b=True, linestyle = 'dashed', linewidth = 1, alpha = 0.4 )
legend = plt.legend(loc=0, prop = {'size':15})
legend.get_frame().set_facecolor(hex_background)

plt.tight_layout()
plt.show()
```

## Plot 4. Intersection of surfaces

**Cylinder**
$$x^{2} + y^{2} = 9$$

<center>or</center>
$$\vec{X}_{\text{cylinder} }(\theta,h) = 3\cos(\theta)\mathbf{\hat{i}} + 3\sin(\theta)\mathbf{\hat{j}} + h\mathbf{\hat{k}}$$

**Plane** 
$$z+x = 5$$

**_Resulting plane and curve from intersection_**

_Plane_
$$\vec{X}_{\text{plane} }(u,v) = 3v\cos(u)\mathbf{\hat{i}} + 3v\sin(u)\mathbf{\hat{j}} + \left( 5-3v\cos(u) \right)\mathbf{\hat{k}}$$

_Curve_
$$\vec{x}(t) = 3\cos(t)\mathbf{\hat{i}} + 3\sin(t)\mathbf{\hat{j}} + \left( 5-3\cos(t) \right)\mathbf{\hat{k}}$$



```python
#cylinder 
#parameters
theta = np.linspace( 0, 2*np.pi, 100 )
h = np.linspace( 0, 9, 100 )

#meshgrid for the parameters
theta, h = np.meshgrid( theta, h )

#parametric equation
X_cyl = 3*np.cos(theta)
Y_cyl = 3*np.sin(theta)
Z_cyl = h

#plane points
u_plane = np.linspace( -5, 5, 100 )
v_plane = np.linspace( -5, 5, 100 )

#meshgrid for plane parameters
u_plane, v_plane = np.meshgrid( u_plane, v_plane )

#plane equation
X_pl = u_plane
Y_pl = v_plane
Z_pl = 5 - u_plane


#vector field
X = np.linspace( -5, 5, 8 )
Y = np.linspace( -5, 5, 8 )
Z = np.linspace( 0, 10, 8 )

#meshgrid for the vector field
X, Y, Z = np.meshgrid ( X, Y, Z )

U = X*Y
V = 2*Z
W = 3*Y

#non interactive plot
%matplotlib inline

#interactive plot - change figsize to a smaller one, like (4,4) and change and comment the "dpi = 200" line
#%matplotlib notebook


#Create figure and the background color
fig = plt.figure( figsize=(10,8),
                 #figsize = (5,5)
                 dpi = 200,
                )
ax = Axes3D(fig)
ax.set_facecolor( hex_background )
ax.w_xaxis.set_pane_color( rgba_planes )
ax.w_yaxis.set_pane_color( rgba_planes )
ax.w_zaxis.set_pane_color( rgba_planes )


#cylinder
ax.plot_surface( X_cyl, Y_cyl, Z_cyl,
                shade = False,
                color = '#45ADA8',
                alpha = 0.70,
               )

#plane
ax.plot_surface( X_pl, Y_pl, Z_pl,
                shade = False,
                color = '#E5FCC2',
                alpha = 0.50,
               )

#vector field
ax.quiver( X, Y, Z, 
          U, V, W, 
          normalize = True,
          length = 0.7,
          lw = 1.3,
          colors = '#9DE0AD',
          label = r'Campo vectorial $\vec{F}(x,y,z) = xy\hat{i} + 2z\hat{j} + 3y\hat{k} $'
         )



#label of our axes
ax.set_xlabel( '$x$', 
              labelpad = 15,
              size = 20
             )
ax.set_ylabel( '$y$', 
              labelpad = 15,
              size = 20
             )
ax.set_zlabel( '$z$', 
              labelpad = 10,
              size = 20
             )

#size of the axes
ax.tick_params( axis = 'x', labelsize = 15 )
ax.tick_params( axis = 'y', labelsize = 15 )
ax.tick_params( axis = 'z', labelsize = 15 )

legend = ax.legend(loc=0, prop = {'size':15})
legend.get_frame().set_facecolor(hex_background)
ax.grid(b = False)

plt.show()
```
